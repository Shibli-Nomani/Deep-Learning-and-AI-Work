# 👽 Tokenization

Tokenization is the process of breaking text into smaller units such as words, phrases, or symbols, known as tokens.

**Purpose of Tokenization in one line**
To segment and organize textual data for further analysis or processing, facilitating language understanding in NLP tasks.


**Alternative of tokenization in online:**
Some cloud-based NLP services and libraries might offer tokenization services as part of their functionalities, allowing users to tokenize text without implementing the process manually. Services like `spaCy`, `NLTK`, or `cloud platforms` like `Google Cloud Natural Language API` or `AWS Comprehend` may offer tokenization features as part of their `NLP toolsets`.

### 👊 word Tokenization
**Tokenized by words**

**`word_tokenize`** and **`sent_tokenize`** functions from NLTK's tokenize module for word and sentence tokenization, respectively.

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
#for punctuation
nltk.download('punkt')
```

```python
sen = "The process of tokenization involves breaking a sequence of text into individual words, symbols, or other meaningful elements."
```

```python
word_token = word_tokenize(sen)
word_token
```

```python
word_token[2:6]
```

### 👊 Sentence Tokenization
**Tokenized by Sentences**

```python
sentences = "Sentence tokenization, an integral aspect of natural language processing, involves breaking down blocks of text into individual sentences using specific punctuation marks like periods, exclamation points, and question marks. Its significance lies in aiding various NLP tasks such as text summarization, sentiment analysis, and machine translation. However, challenges can arise in sentence tokenization due to diverse writing styles, varied punctuation usage, and the presence of unconventional sentence structures across different languages."
```

```python
sen_token = sent_tokenize(sentences)
sen_token
```

```python
sen_token[0]
```

# 👽 Stemming
👊**Chopping Out**

Stemming is the process of reducing words to their base or root form by removing affixes, which may not always result in valid words but aims to cut off prefixes or suffixes to obtain the word's core meaning.

 **`PorterStemmer class`** from NLTK's stem module, used for stemming words using the Porter stemming algorithm

```python
from nltk.stem import PorterStemmer
```

```python
en_steam = PorterStemmer()
```

```python
sen_define = "Stemming is the process of reducing words to their root or base form, often by removing affixes, to improve information retrieval in natural language processing and text mining. Lemmatization is the process of obtaining the base or dictionary form of a word to normalize variations and derive its meaningful root, considering the context within natural language processing."
```

```python
word_token = word_tokenize(sen_define)
word_token
```

```python
stem_sen = [en_steam.stem(i) for i in word_token]
stem_sen
```

# 👽Lemmatization

👊**Context Meaning of Word**

Lemmatization is the process of obtaining the base or dictionary form of a word (known as a lemma) to normalize variations, considering the context, and producing valid words.

**`nltk.download('wordnet'):`** This command downloads the WordNet dataset, which is a lexical database of English words and their semantic relationships, including synonyms, antonyms, hypernyms, hyponyms, etc. WordNet is commonly used in tasks like word sense disambiguation, similarity measures, and semantic analysis.

**`nltk.download('omw-1.4'):`** This command downloads the Open Multilingual WordNet (OMW) dataset version 1.4. OMW extends WordNet by providing multilingual synsets, offering word translations and relations in various languages linked to WordNet synsets.

```python
from nltk.stem import WordNetLemmatizer
import nltk
#from nltk.corpus import wordnet
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
sen_for_lemma = "Stemming is the process of reducing words to their root or base form, often by removing affixes, to improve information retrieval in natural language processing and text mining. Lemmatization is the process of obtaining the base or dictionary form of a word to normalize variations and derive its meaningful root, considering the context within natural language processing."

word_token = word_tokenize(sen_for_lemma)
word_token
```

```python
en_lemma = WordNetLemmatizer()

word_lemma = [en_lemma.lemmatize(i) for i in word_token]
word_lemma
```

# 😆 Comparing Stemming vs Lemmatization

In summary, stemming is faster and less resource-intensive but might sacrifice linguistic accuracy, producing non-words occasionally. Lemmatization, on the other hand, is slower but provides accurate base words, making it more suitable for applications requiring grammatical correctness and semantic understanding,

```python
stem_sen[15:25]
```

```python
word_lemma[15:25]
```

# 👽 POS TAGGING

POS tagging, or Part-of-Speech tagging, is the process of assigning grammatical categories (such as noun, verb, adjective) to words in a text corpus.

👊**Importance of POS Tagging:**

POS tagging is important in NLP for understanding the syntactic structure of sentences, aiding in grammar analysis, and facilitating accurate language understanding for various natural language processing tasks.

**`nltk.download("averaged_perceptron_tagger")`** is for part-of-speech tagging using the averaged perceptron algorithm.

**The averaged perceptron algorithm** predicts the parts of speech (POS) of words in text by learning patterns from labeled data and adjusting features to make accurate predictions.

```python
nltk.download("averaged_perceptron_tagger")
```

```python
pos_tagg = nltk.pos_tag(word_token)
```

```python
pos_tagg
```

# 👽 NER (Name Entity Recognation)
NER stands for Named Entity Recognition, a process in natural language processing (NLP) that identifies and categorizes named entities such as names of persons, organizations, locations, dates, and more within text data.

**`maxent_ne_chunker`** is used for Named Entity Recognition (NER).

**`ne_chunk`** stands for "Named Entity Chunker". This function is specifically designed to recognize and label named entities in text

```python
nltk.download('words')
nltk.download("maxent_ne_chunker")
```

```python
#! pip install svgling
```

```python
sen_sample = "Steve Jobs was the CEO of Apple Inc. And Sam Altman is the CEO of OpenAI"

word_tokens = word_tokenize(sen_sample)
pos_taggs = nltk.pos_tag(word_tokens)
en_chunks = nltk.ne_chunk(pos_taggs)
en_chunks
```
